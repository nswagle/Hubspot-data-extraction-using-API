```python
import requests
import pandas as pd
import json

url = "https://api.hubapi.com/crm/v3/objects/contacts"
YOUR_ACCESS_TOKEN = "pat-eu1-1c90cffe-64af-44f1-b0e3-afe6a2041f32"
headers = {
  'content-type': 'application/json',
  'authorization': 'Bearer %s' % YOUR_ACCESS_TOKEN
}

response = requests.request("GET", url, headers=headers)


js2 = json.loads(response.text)

allResults = list()
counter = 0

while('paging' in js2):
    allResults.extend(js2['results'])
    url = js2['paging']['next']['link']
    print(url)
    response = requests.request("GET", url, headers=headers)
    js2 = json.loads(response.text)
    counter = counter + 1
    
out_file = open("contactsNew.json", "w")
  
json.dump(allResults, out_file, indent = 6)
  
out_file.close()
```
